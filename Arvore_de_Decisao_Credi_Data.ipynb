```python
# Importando minha bibilioteca para Arvore de decisão 
# from sklearn.tree import DecisionTreeClassifier -- >  Classficador basedo em arvore de decisão 
from sklearn.tree import DecisionTreeClassifier
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
# Carregando meus dados
import pickle
with open('credit.pkl', 'rb') as f:
  x_credit_treinamento, y_credit_treinamento, x_credit_teste, y_credit_teste = pickle.load(f)
```

```python
# Vendo o shape da minha base treinamento
x_credit_treinamento.shape, y_credit_treinamento.shape
```

```python
# Verificando o tamanho da base de teste 
x_credit_teste.shape , y_credit_teste.shape
```

```python
# Criando minha arvore de decisão sendo assim o algoritimo e treinado

arvore_credit = DecisionTreeClassifier(criterion="entropy", random_state= 0)
arvore_credit.fit(x_credit_treinamento, y_credit_treinamento)
```

```python
# Criando minha variavel de previsores
previsores = arvore_credit.predict(x_credit_teste)
previsores

# Sendo a classe 1 pessoa nao pagou o emprestimo
# Sendo a classe 0 pessoa pagou o emprestimo
```

```python
# Comparando os valores com minha base real
y_credit_teste
```

```python
# Importando minha biblioteca para a metragem
from sklearn.metrics import accuracy_score, classification_report
```

```python
# Passando as variaveis para saber quanto de acerto meu algoritimom fez
accuracy_score(y_credit_teste, previsores)

# Sendo assim com 98 % de acerto ja no naive_bayes temos o acerto de 93 % 
```

```python
# Criando minha matriz de confusão 
# importando minha bibliotca
from yellowbrick.classifier import ConfusionMatrix

# Criando minha matriz de confusão
cm = ConfusionMatrix(arvore_credit)
cm.fit(x_credit_treinamento, y_credit_treinamento)
cm.score(x_credit_teste, y_credit_teste)

# Sendo assim na classe 0 430 clietes pagam e apenas 6 foram classificados errados
# Sendo assim na classe 1 61 clietes não pagam e apenas 3 foram classificados errados
```

```python
# Vizualizando meu relatorio de classificação
print(classification_report(y_credit_teste, previsores))
```
